```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw_Data.csv")
df = df[df.columns[1:]]

df.columns = list(map(lambda  x : "V"+str(x), list(range(1,16))))
display(df)
df.to_csv("Answers.csv", index=False)
n_values = df.size
n_to_remove = n_values // 2

# Create a mask to randomly select values to remove
mask = np.random.choice([True, False], size=n_values, replace=True, p=[0.5, 0.5])
mask = mask.reshape(df.shape)
#print(len(mask), mask)

# Replace the selected values with np.nan
df[mask] = np.nan

display(df)
df.to_csv("Training_set.csv", index=False)
```
